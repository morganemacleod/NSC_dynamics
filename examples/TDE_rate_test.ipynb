## Test case: Compare to the TDE rate computed by Wang & Merritt 2004

This is applicable for a isothermal ($\rho(r) \propto r^{-2}$) galaxy, for sun-like stars. Further, the assumption is that twice the black hole mass of stars are contained within a radius $r_m = r_h$. 

We capture these with the following settings (commented below):

```python
# COMPARE TO WM2004 SETTINGS
n=nsc_model(Mbh=1.e6*c.msun,  # black hole mass of 1e6 solar masses
            ms=c.msun,        # stellar mass of 1 solar mass
            gamma=2.,         # rho(r) \propto r^-gamma, here we set gamma=2
            rm_o_rh=1.,       # r_m / r_h -- the ratio of the radius enclosing a certain mass of stars to the sphere of influence radius 
            mrm_o_mbh=2.)     # mass in stars / Mbh within r_m

n.test_plot_TDE_rate()
```
